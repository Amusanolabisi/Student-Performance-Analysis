```python
## import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## load the csv file
df = pd.read_csv("student_performance.csv")
df
```

```python
## make a copy of the file
student_performance = df.copy()
```

```python
## view the first five rows
student_performance.head()
```

```python
## retrieve the info about the data
student_performance.info()
```

```python
## retrieve the data statistics information
student_performance.describe()
```

```python
# check the data types
student_performance.dtypes
```

```python
## draw random row of the data
student_performance.sample()
```

```python
## check the count of null values in the columns
student_performance.isnull().sum()
```

```python
## retrieve the shape of the data
student_performance.shape
```

```python
## drop nas
student_performance = student_performance.dropna(subset=['total_score', 'attendance_pct'])
```

```python
## Calculate pass rate per subject and visualize as a horizontal bar chart
pass_rate = student_performance.groupby("course")["passed"].apply(lambda x: (x == "Yes").mean() * 100).sort_values()
pass_rate.plot(kind="barh", figsize= (12,8))
plt.title("Pass Rate by Subject (%)")
plt.xlabel("pass rate")
plt.show()
```

```python
## Box plot: Score distribution by school type (Public vs Private)
sns.boxplot(data=student_performance, x='school_type', y='total_score')
plt.title("Score Distribution by School Type")
plt.show()
```

```python
## Scatter plot: Attendance % vs Total Score with a regression line
sns.regplot(data=student_performance, x='attendance_pct', y='total_score', scatter_kws={'alpha':0.5})
plt.title("Attendance vs Total Score")
plt.show()
```

```python
## Bar chart: Average score by LGA
avg_lga = student_performance.groupby('lga')['total_score'].mean().sort_values(ascending=False)
avg_lga.plot(kind='bar', figsize=(10,5))
plt.title("Average Score by LGA")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```

```python
## Grouped bar chart: Pass rate by gender across subjects
pass_gender = student_performance.groupby(['course', 'gender'])['passed'].apply(lambda x: (x=='Yes').mean()*100).reset_index()

sns.barplot(data=pass_gender, x='course', y='passed', hue='gender')
plt.title("Pass Rate by Gender Across Subjects")
plt.xticks(rotation=45)
plt.show()
```

```python
## Heatmap: Average score by subject and academic year
pivot = student_performance.pivot_table(values='total_score',
                       index='course',
                       columns='academic_year',
                       aggfunc='mean')

sns.heatmap(pivot, annot=True, cmap='coolwarm')
plt.title("Average Score by Subject and Year")
plt.show()
```

```python
## Bar chart: Score comparison by parental education level
edu_avg = df.groupby('parent_education')['total_score'].mean().sort_values()

edu_avg.plot(kind='bar', figsize=(8,5))
plt.title("Average Score by Parental Education")
plt.ylabel("Average Score")
plt.show()
```

```python
## Line chart: Failure rate trend across the three academic years
fail_trend = student_performance.groupby('academic_year')['passed'].apply(lambda x: (x=='No').mean()*100)

fail_trend.plot(marker='o')
plt.title("Failure Rate Trend Over Years")
plt.ylabel("Failure Rate (%)")
plt.show()
```

```python
## Correlation matrix: Attendance, study hours, CA score, exam score
corr_cols = ['attendance_pct', 'ca_score', 'exam_score', 'total_score']

corr = student_performance[corr_cols].corr()

sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```

```python
## Flag at-risk students: attendance below 60% AND total score below 45
student_performance['risk_flag'] = 'Safe'

student_performance.loc[
    (student_performance['attendance_pct'] < 60) & (student_performance['total_score'] < 45),
    'risk_flag'
] = 'At Risk'
student_performance[['student_id', 'attendance_pct', 'total_score', 'risk_flag']].head()
```
